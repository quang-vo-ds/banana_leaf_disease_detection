## Initial Setup

```python
import os
import glob
import pickle
import numpy as np
import pandas as pd
import csv

from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/Vin_ML_Course/Final_Project
root_dir = os.getcwd()
```

## Create CSV Data

```python
class_int_to_name = {
    0: "Healthy",
    1: "aphids",
    2: "bacterial_soft_rot",
    3: "bacterial_wilt",
    4: "black_segatoka",
    5: "pseudostem_weevil",
    6: "Xanthomonas_wilt"
}
class_name_to_int = {v:k for k,v in class_int_to_name.items()}
print("Map class name to integer: ", class_name_to_int)
```

```python
csv_data = {'id': [], 'class': [], 'path': []}
for k, v in class_int_to_name.items():
    class_path = os.path.join(root_dir, "data", "Banana___"+v)
    for f in glob.glob(class_path + '/*'):
        f_name = f[f.rfind("/")+1:]

        # id and class
        csv_data['id'].append(f_name)
        csv_data['class'].append(v)
        csv_data['path'].append(f)

csv_data = pd.DataFrame(csv_data)
csv_data = csv_data.dropna()
csv_data
```

## Save file

```python
# Create processsed_data directory
save_dir = os.path.join(root_dir, "output/processed_data")
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Save csv_data,
csv_data.to_csv(os.path.join(save_dir, "csv_data.csv"), index=False)

with open(os.path.join(save_dir,'class_int_to_name.pkl'), 'wb') as fp:
    pickle.dump(class_int_to_name, fp)

with open(os.path.join(save_dir, 'class_name_to_int.pkl'), 'wb') as fp:
    pickle.dump(class_name_to_int, fp)
```
